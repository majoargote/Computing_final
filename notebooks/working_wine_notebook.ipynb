```python
# Check interaction between different features and quality

def corr_X_y(X, y, methods=('pearson','spearman')):
    # Ensure pandas objects
    X = pd.DataFrame(X).copy()
    y = pd.Series(y).squeeze()

    # Force numeric (turn 'average' etc. into NaN)
    Xn = X.apply(pd.to_numeric, errors='coerce')
    yn = pd.to_numeric(y, errors='coerce')

    # Keep rows where y is present, and align indices
    mask = yn.notna()
    Xn = Xn.loc[mask]
    yn = yn.loc[mask]

    out = pd.DataFrame({m: Xn.corrwith(yn, method=m) for m in methods})
    out['abs_spearman'] = out['spearman'].abs()
    return out.sort_values('abs_spearman', ascending=False)

corr_table = corr_X_y(X, y)
corr_table
```
